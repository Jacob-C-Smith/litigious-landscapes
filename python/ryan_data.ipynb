```python
# Defines functions for reading and writing data

# Imports
import csv
from datetime import date, datetime, timedelta
import random

# Load a CSV file from a path
def read_table ( path : str ):
    
    # Initialized data
    table : list = [ ]

    # Open the CSV file
    with open(path) as csvfile:

        # Open a CSV file reader
        csv_reader = csv.reader(csvfile)
        
        # Dump the table
        for r in csv_reader:
            
            # Store the row
            table.append(r)
    
    # Success
    return ( table, len(table), len(table[0]) )

# Write a table to a CSV file
def write_table ( path : str, table : list, column_headers : list ):
    
    # Open the CSV file
    with open(path, 'w', newline='') as csvfile:

        # Open a CSV file reader
        csv_writer = csv.writer(csvfile)
        
        # Write the column headers
        csv_writer.writerow(column_headers)

        # Dump the table
        csv_writer.writerows(table)
        
    # Success
    return 
```
